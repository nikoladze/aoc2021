```python
from solver import find_basins, parse
import numpy as np
import matplotlib.pyplot as plt
from collections import deque
```

```python
data = parse(open("input.txt").read())
```

```python
basins = find_basins(data)
```

```python
grid = np.array(data)
grid[:] = -1
for i, basin in enumerate(np.random.permutation(basins)):
    for x, y in basin:
        grid[y, x] = i
```

```python
plt.figure(figsize=(12, 12))
plt.imshow(grid, cmap="turbo")
```
